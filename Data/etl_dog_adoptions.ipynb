### Load in csv file

```python
import pandas as pd
df = pd.read_csv('source_data\dog_adoptions2020_2021.csv')
df.head()
```

### Split Date and Time up, change to datatime and create a column for days until adoption

```python
df['adopt_date']=df['status_changed_at'].str.split('T').str[0]
df['post_date']=df['published_at'].str.split('T').str[0]
df.head()
```

```python
df['adopt_date'] = pd.to_datetime(df['adopt_date'])
df['post_date'] = pd.to_datetime(df['post_date'])
df['adoption_days']=(df['adopt_date']-df['post_date']).dt.days
df
```

### There are many breeds that have only a couple of instances so we are changing breeds with under 100 instances to 'Speciality'

```python
df['breed_count'] = df['primary_breed'].map(df['primary_breed'].value_counts())
df['primary_breed'].nunique()
```

```python
df['primary_breed'].loc[(df['breed_count'] < 250)] = 'Speciality'
df['primary_breed'].nunique()
```

### We are taking null values of 'Coat', 'gw_children', 'gw_dogs', 'gw_cats' and replacing with 'Unanswered' This is because we will access the other labels and will consider the unanswered as a label

### We are then taking the tags, description and photo columns and creating a new column that is either Ture they entered data into those or False they left them blank, we are not exploring what was entered into these columns but just if the fact that there was something in them if it impacted adoption time

```python
df['coat'] = df['coat'].fillna('Unanswered')
df['primary_color'] = df['primary_color'].fillna('Unanswered')
df['gw_children'] = df['gw_children'].fillna('Unanswered')
df['gw_dogs'] = df['gw_dogs'].fillna('Unanswered')
df['gw_cats'] = df['gw_cats'].fillna('Unanswered')
```
