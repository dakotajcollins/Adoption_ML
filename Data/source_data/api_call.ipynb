```python
def append_list(animals):

    for i in range(1,len(animals)):

        id = animals[i]['id']
        name = animals[i]['name']
        url = animals[i]['url']
        type = animals[i]['type']
        primary_breed = animals[i]['breeds']['primary']
        secondary_breed = animals[i]['breeds']['secondary']
        mixed_breed = animals[i]['breeds']['mixed']
        unknown_breed = animals[i]['breeds']['unknown']
        primary_color = animals[i]['colors']['primary']
        secondary_color = animals[i]['colors']['secondary']
        age = animals[i]['age']
        gender = animals[i]['gender']
        size = animals[i]['size']
        coat = animals[i]['coat']
        house_trained = animals[i]['attributes']['house_trained']
        spayed_neutered = animals[i]['attributes']['spayed_neutered']
        special_needs = animals[i]['attributes']['special_needs']
        shots_current = animals[i]['attributes']['shots_current']
        gw_children = animals[i]['environment']['children']
        gw_dogs = animals[i]['environment']['dogs']
        gw_cats = animals[i]['environment']['cats']
        tags = animals[i]['tags']
        description = animals[i]['description']
        photo_urls = animals[i]['photos']
        status = animals[i]['status']
        status_changed_at = animals[i]['status_changed_at']
        published_at = animals[i]['published_at']
        city = animals[i]['contact']['address']['city']
        state = animals[i]['contact']['address']['state']
        postcode = animals[i]['contact']['address']['postcode']
        country = animals[i]['contact']['address']['country']

        dict = {'id': id, 'name': name, 'url': url, 'type': type, 'primary_breed' : primary_breed,
                'seondary_breed' : secondary_breed, 'mixed_breed' : mixed_breed, 'unknown_breed' : unknown_breed,
                'primary_color' : primary_color, 'secondary_color' : secondary_color, 'age' : age, 'gender' : gender,
                'size' : size, 'coat' : coat, 'house_trained' : house_trained, 'spayed_neutered' : spayed_neutered,
                'special_needs' : special_needs, 'shots_current' : shots_current, 'gw_children' : gw_children,
                'gw_dogs' : gw_dogs, 'gw_cats' : gw_cats, 'tags' : tags, 'description' : description, 'photo_urls' : photo_urls,
                'status' : status, 'status_changed_at' : status_changed_at, 'published_at' : published_at, 'city' : city,
                'state' : state, 'postcode' : postcode, 'country' : country     
               }

        list.append(dict)
```

```python
for j in range(1,5000):

    headers = {
         'Authorization': auth_string,
    }

    params = (
        ('type', 'dog'),('status', 'adopted'),('page', j),('before', '2020-11-23T14:22:15+0000')
    )
    print(j)
    response = requests.get('https://api.petfinder.com/v2/animals', headers=headers, params=params)

    animals=response.json()['animals']

    append_list(animals)
```
